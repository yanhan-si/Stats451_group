```python
import numpy as np
import pandas as pd
# import csv
from datetime import datetime

import seaborn as sns
color = sns.color_palette()

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
bitcoin = pd.read_csv('bitcoin.csv')
bitcoin.columns
```

### distributions

```python
plt.figure()
sns.distplot(bitcoin['Close'], kde=False, label='closing price') 
# sns.distplot(bitcoin['Open'], kde=False, label='opening price')
plt.xlabel('closing price')
plt.title("Distribution of closing price of Bitcoin")
plt.legend(loc='best')
# plt.savefig("dist_bit.png")

plt.figure()
sns.distplot(bitcoin['ret'], kde=False, label='daily return')
plt.xlabel('daily return')
plt.title("Distribution of daily return of Bitcoin")
plt.legend(loc='best')
# plt.savefig("dist_ret.png")

plt.figure()
sns.distplot(bitcoin['log_return'], kde=False, label='log daily return')
plt.xlabel('log daily return')
plt.title("Distribution of log daily return of Bitcoin")
plt.legend(loc='best')
# plt.savefig("dist_logret.png")
plt.show()
```

### price over time

```python
xdate = [datetime.strptime(d,'%Y-%m-%d').date() for d in bitcoin.Date]

plt.plot(xdate, bitcoin.Close, label='closing price')
plt.gcf().autofmt_xdate()
plt.xlabel("Date")
plt.ylabel("Bitcoin Price")
plt.title("Bitcoin Price overtime")
plt.legend(loc='best')
# plt.savefig("overtime_price.png")

plt.show()
```

```python
xdate = [datetime.strptime(d,'%Y-%m-%d').date() for d in bitcoin.Date]
bitcoin['moving_avg'] =  bitcoin['Close'].rolling(window=10).mean()

plt.plot(xdate, bitcoin.Close, label='closing price')
plt.plot(xdate, bitcoin.moving_avg, label='moving average - 15')
plt.gcf().autofmt_xdate()
plt.xlabel("Date")
plt.ylabel("Bitcoin Price")
plt.title("Bitcoin Price overtime")
plt.legend(loc='best')
# plt.savefig("overtime_price_avg.png")

plt.show()
```

```python
plt.plot(xdate, bitcoin.ret, label='daily return')
plt.gcf().autofmt_xdate()
plt.xlabel("Date")
plt.ylabel("daily return")
plt.title("Variation overtime")
plt.legend(loc='best')
plt.savefig("overtime_var.png")

plt.show()
```

```python
plt.plot(xdate, bitcoin.log_return, label='daily return')
plt.gcf().autofmt_xdate()
plt.xlabel("Date")
plt.ylabel("log daily return")
plt.title("Log variation overtime")
plt.legend(loc='best')
plt.savefig("overtime_logvar.png")

plt.show()
```
